# SQL Select Lab - Challenges 1 and 2
This notebook connects to `publications.sqlite` and solves challenges 1 and 2.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('publications.sqlite')
conn
```

```python
# Quick row-count check target used by both challenges
target_rows = pd.read_sql_query('SELECT COUNT(*) AS cnt FROM titleauthor', conn).iloc[0]['cnt']
target_rows
```

## Challenge 1 - Who Have Published What At Where?

```python
challenge_1_query = '''
SELECT
    a.au_id AS "AUTHOR ID",
    a.au_lname AS "LAST NAME",
    a.au_fname AS "FIRST NAME",
    t.title AS "TITLE",
    p.pub_name AS "PUBLISHER"
FROM titleauthor ta
JOIN authors a
    ON ta.au_id = a.au_id
JOIN titles t
    ON ta.title_id = t.title_id
JOIN publishers p
    ON t.pub_id = p.pub_id
ORDER BY a.au_id, t.title;
'''

challenge_1_df = pd.read_sql_query(challenge_1_query, conn)
print(f'Challenge 1 rows: {len(challenge_1_df)} | Expected: {target_rows}')
challenge_1_df.head(20)
```

## Challenge 2 - Who Have Published How Many At Where?

```python
challenge_2_query = '''
SELECT
    a.au_id AS "AUTHOR ID",
    a.au_lname AS "LAST NAME",
    a.au_fname AS "FIRST NAME",
    p.pub_name AS "PUBLISHER",
    COUNT(*) AS "TITLE COUNT"
FROM titleauthor ta
JOIN authors a
    ON ta.au_id = a.au_id
JOIN titles t
    ON ta.title_id = t.title_id
JOIN publishers p
    ON t.pub_id = p.pub_id
GROUP BY a.au_id, a.au_lname, a.au_fname, p.pub_name
ORDER BY a.au_id, p.pub_name;
'''

challenge_2_df = pd.read_sql_query(challenge_2_query, conn)
print(f'Sum of TITLE COUNT: {challenge_2_df["TITLE COUNT"].sum()} | Expected: {target_rows}')
challenge_2_df.head(20)
```

## Challenge 3 - Best Selling Authors
Top 3 authors by total quantity sold.

```python
challenge_3_query = '''
SELECT
    a.au_id AS "AUTHOR ID",
    a.au_lname AS "LAST NAME",
    a.au_fname AS "FIRST NAME",
    SUM(s.qty) AS "TOTAL"
FROM authors a
JOIN titleauthor ta
    ON a.au_id = ta.au_id
JOIN sales s
    ON ta.title_id = s.title_id
GROUP BY a.au_id, a.au_lname, a.au_fname
ORDER BY "TOTAL" DESC
LIMIT 3;
'''

challenge_3_df = pd.read_sql_query(challenge_3_query, conn)
challenge_3_df
```

## Challenge 4 - Best Selling Authors Ranking
Show all authors (including those with 0 sales), ordered by total sold descending.

```python
challenge_4_query = '''
SELECT
    a.au_id AS "AUTHOR ID",
    a.au_lname AS "LAST NAME",
    a.au_fname AS "FIRST NAME",
    COALESCE(SUM(s.qty), 0) AS "TOTAL"
FROM authors a
LEFT JOIN titleauthor ta
    ON a.au_id = ta.au_id
LEFT JOIN sales s
    ON ta.title_id = s.title_id
GROUP BY a.au_id, a.au_lname, a.au_fname
ORDER BY "TOTAL" DESC, a.au_id;
'''

challenge_4_df = pd.read_sql_query(challenge_4_query, conn)
print(f'Rows returned: {len(challenge_4_df)} (should be all authors)')
challenge_4_df
```
